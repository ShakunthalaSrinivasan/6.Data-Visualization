```python
sb.violinplot(x='degree_t',y='salary',data=dataset,hue='gender')
plt.ylim(0,1000000)
```

Takeaways from the violin plot between degree courses and salary

1. Student from the department 'commerce & Management' earn the highest salary among all.
2. Majority of the Male students from the department 'Science & Technology' earns between 2.5 to 4 lakhs. 
   Majority of the female students from the department 'Science & Technology' earns between 2.2 to 3 lakhs
3. Majority of the male students fron the department 'commerce & Management' earns between 2.3 to 3 lakhs.
   Majority of the female students fron the department 'commerce & Management' earns between 2 to 3 lakhs.
4. Only one male student from the department 'others' is placed.
   Majority of the female students fron the department 'others' earns between 2.5 to 3.5 lakhs.
5. Male students from the department 'Science & Technology' and 'Commerce & Management' earn more salary than the female students.
6. White dot in the plot represents the median value in each category.


```python
sb.boxplot(x='gender',y='degree_p',data=dataset)
plt.ylim(0,100)
```

Takeaways from the graph:
1. Female students secured high marks in degree than the male students.
2. The maximum and minimum range of degree marks of male students lies between 50 and 80.
3. The maximum and minimum range of degree marks of female students lies between 55 and 85. Few outliers are also present in the female category.
4. The median value of male students degree marks is 65 which means 50% of the male students have secured above 65 and 50% secured below 65 marks.
5. The median value of female students degree marks is 70 which means 50% of the female students have secured above 70 and 50% secured below 70 marks.

```python
data=sb.load_dataset('exercise')
data
data.to_csv("factorplot.csv")
```

```python
sb.catplot(x='time',y='pulse',hue='kind',kind='violin',col='diet',data=data)
```

Take Aways from the graph
1. In general, the pulse rate in all kinds is on higher side for the no fat people.
   The pulse rate spikes more than 120 and 100 when the people are in the running phase and walking phase respectively.
2. The white line inside the bar represents the median value which means 50% of the people's pulse rate is lesser than the median value and 50% is more than the median value in the particular category.  
3. In the rest phase:
   * The pulse rate of the no fat people ranges between 82 to 100.
   * The pulse rate of the low fat people ranges between 80 to 96.
4. In the Walking phase:
   * The pulse rate of the no fat people ranges between 82 to 110.
   * The pulse rate of the low fat people ranges between 83 to 103.
5. In the Running phase
   * The pulse rate of the no fat people ranges between 95 to 145.
   * The pulse rate of the low fat people ranges between 90 to 130.
